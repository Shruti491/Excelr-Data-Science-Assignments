# Interview Questions


## 1.What is Normalization & Standardization and how is it helpful?

 Normalization: Scales data to a range 0 to 1 and Useful when features have different scales but a fixed range is needed (e.g., pixel data).

Standardization: Scales data to have a mean of 0 and a standard deviation of 1 and Best when data follows a normal distribution or for algorithms sensitive to variance (e.g., Logistic Regression).

It helpful for:

1.Improves model performance by balancing feature scales.

2.Speeds up optimization (e.g., gradient descent).

3.Prevents dominance of large-scale features.

4.Enhances fairness and interpretability.

## 2.What techniques can be used to address multicollinearity in multiple linear regression?

1.Remove Highly Correlated Predictors i.e. Drop one of the variables that show high correlation.

2.Variance Inflation Factor (VIF) Identify and remove predictors with high VIF values (typically > 5 or 10).

3.Regularization: Use techniques like Ridge Regression or Lasso Regression to penalize and reduce the impact of multicollinearity.

4.Principal Component Analysis (PCA): Transform correlated predictors into a smaller set of uncorrelated components.

5.Combine Variables: Combine correlated variables into a single feature (e.g., averaging or summing them).

6.Increase Sample Size: Collect more data to reduce the impact of multicollinearity on model estimation.
