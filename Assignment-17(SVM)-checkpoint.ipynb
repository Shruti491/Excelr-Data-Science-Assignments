```python
x = df.drop(columns=['class','cap_shape','cap_surface','cap_color','bruises','odor','gill_attachment','stalk_root','stalk_surface_above_ring','stalk_surface_below_ring','stalk_color_above_ring','gill_spacing','gill_size','gill_color','stalk_shape','stalk_color_below_ring','veil_type','veil_color','ring_number','ring_type','spore_print_color','population','habitat'])
y = df['class']
```

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest = train_test_split(x,y, test_size=0.2, random_state=1)
svc = SVC()

svc.fit(xtrain,ytrain)
ypred = svc.predict(xtest)
print(svc.score(xtrain,ytrain))
print(svc.score(xtest,ytest))
```
